```python
from rift_zones import *

# Sliced sine wave just for a quick test
num_points = 100
x_start = 0
x_end = 15000
amplitude = 3000
frequency = 0.001

x_values = np.linspace(x_start, x_end, num_points) - 2000
y_values = 1500 + amplitude * np.sin(frequency * x_values) + amplitude

pos_df = pd.DataFrame({
    'pos_x': x_values,
    'pos_y': y_values,
})

pos_df['pos_x'] = pos_df['pos_x'].astype(int)
pos_df['pos_y'] = pos_df['pos_y'].astype(int)

pos_df = pos_df.iloc[60:72]

show_zone(pos_df=pos_df, fig_size=(10, 10))
```
